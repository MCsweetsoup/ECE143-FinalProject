```python
#Food Security Status of U.S. Households with Children in 2020
import matplotlib.pyplot as plt
from palettable.colorbrewer.qualitative import Pastel1_7
 
# create data
names = ['85.2 percent of households with children were food secure', 
         '7.2 percent of households with children, only adults were food insecure', 
         'low food security among children - 6.8%', 'very low food security among children - 0.8%,']
size = [85.2 ,7.2,6.8, 0.8]
 
# Create a circle at the center of the plot
my_circle = plt.Circle( (0,0), 0.7, color='white')
label = ['85.2% of households with children were food secure', 
         '7.2%  of households with children, only adults were food insecure', 
         'low food security among children - 6.8%', 
         'very low food security among children - 0.8%,']
from palettable.colorbrewer.qualitative import Pastel1_7
plt.figure(figsize=(8,8))
plt.pie(size, colors=Pastel1_7.hex_colors)
p = plt.gcf()
p.legend(label, loc="center left")
p.gca().add_artist(my_circle)

# Show the graph
# plt.legend(loc="upper right")
plt.show()
```

```python
# Food Insecurity by Household Characteristics
import pandas as pd
data = pd.read_excel('/content/drive/My Drive/143 project/insecurity2020.xlsx')
```

```python
data
```

```python
import numpy as np
my_range = range(1, 36)
plt.figure(figsize = (8,8))
my_color=np.where(data['Category']=='All households', 'orange', 'skyblue')
my_size=np.where(data['Category']=='All households', 70, 30)
plt.hlines(y = my_range, xmin = 0, xmax=data['Percent'], color = my_color)
plt.scatter(data['Percent'], my_range, color = my_color, s = my_size, alpha = 1)

# Add titles and axis names
plt.yticks(my_range, data['Category'])
plt.title("Food Insecurity by Household Characteristics")
plt.xlabel('Percentage')
plt.ylabel('Household Characteristics')

# Show the plot
plt.show()
```
